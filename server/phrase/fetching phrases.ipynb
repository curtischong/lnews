```python
import requests as req
from bs4 import BeautifulSoup
```

```python
resp = req.request(method='GET', url="https://www.phrases.org.uk/meanings/phrases-and-sayings-list.html")
print(resp.text)
```

```python
f = open("all_phrases.txt", "a")

soup = BeautifulSoup(resp.text, 'html.parser')
mydivs = soup.find_all("a",href=True)
for div in mydivs:
    f.write(div['href']  + ";" + div.text.strip() + "\n")
f.close()
```
