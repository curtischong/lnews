```python
import requests as req
from bs4 import BeautifulSoup
```

```python
resp = req.request(method='GET', url="https://www.phrases.org.uk/meanings/phrases-and-sayings-list.html")
print(resp.text)
```

```python
soup = BeautifulSoup(resp.text, 'html.parser')
mydivs = soup.findAll("p", {"class": "phrase-list"})
```

```python
f = open("all_phrases.txt", "a")
for div in mydivs:
    links = div.find_all('a', href=True)
    if(len(links) > 0):
        link = links[0]['href']
        divText = div.text.strip()
        if(divText != ""):
            f.write(link + ";" + divText + "\n")
f.close()
```
